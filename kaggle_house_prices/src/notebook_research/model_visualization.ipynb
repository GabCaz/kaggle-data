# Model Visualization

```python
from matplotlib import pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from kaggle_house_prices.src.features import get_x_y, load_data
from kaggle_house_prices.src.model_pipeline import get_model_pipeline


df_train, df_test = load_data()
X, y = get_x_y(data=df_train)

# look into one example of splitting the data
df_train_train, df_train_valid = train_test_split(
    df_train, test_size=0.2, random_state=42
)
X_train_train, y_train_train = get_x_y(data=df_train_train)
X_train_valid, y_train_valid = get_x_y(data=df_train_valid)

full_model = get_model_pipeline()

full_model.fit(X_train_train, y_train_train)
y_pred_valid = full_model.predict(X_train_valid)
plt.scatter(np.exp(y_train_valid), np.exp(y_pred_valid), alpha=0.2)
plt.title("Predicted vs. True values")
plt.show()
```
